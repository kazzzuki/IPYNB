```python
import pandas as pd
import matplotlib.pyplot as plt
import os

from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/DSP-OGATA/VoltagevsTime.xlsx'
df = pd.read_excel(file_path)

print(df.head(10))
```

### **1. What is the average value for voltage? What is the average and standard deviation?**

```python
aveVolt = df['Amplitude (V)'].mean()
stdVolt = df['Amplitude (V)'].std()

print("Voltage Average:", aveVolt)
print("Voltage Standard Deviation:", stdVolt)
```

### **2. Plot the values of voltage with respect to time. What do you observe?**

```python
plt.figure(figsize=(15, 5))
plt.plot(df['Time (sec)'], df['Amplitude (V)'], label='Voltage vs Time', color='blue')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (V)')
plt.title('Voltage vs Time')
plt.grid(True)
plt.legend()
plt.show()
```

**Observation**:
* I observe that the voltage signal fluctuates over time. There is no constant voltage, and the data seems to go up and down between higher and lower values. This indicates that there is no stability and it suggests the presence of noise or irregularities in the system.

### **3. What is the probability of voltage to have 220 V?**

```python
NumCount220V = (df['Amplitude (V)'] == 220).sum()
totalSample = len(df)

probVolt = NumCount220V / totalSample

print("Probability of Voltage to have 220V:", probVolt)
```

### **4. Aside from the average value, find the other central tendency values.**

```python
medVolt = df['Amplitude (V)'].median()
modVolt = df['Amplitude (V)'].mode()[0]

print("Voltage Median:", medVolt)
print("Voltage Mode:", modVolt)
```

### **5. Plot the histogram for the first 500 samples, then 1000 samples, then 2000 samples, then 5000 samples. What do you observe?**

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

sampleSizes = [500, 1000, 2000, 5000]

for i, ax in enumerate(axs.flat):
    size = sampleSizes[i]
    ax.hist(df['Amplitude (V)'][:size], bins=20, alpha=0.7, color='blue')
    ax.set_title(f'Histogram for First {size} Samples')
    ax.set_xlabel('Amplitude (V)')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### **6. What can you conclude on these graphs?**

*   As the sample size increases from 500 to 1000 to 2000 and to 5000, the distribution of voltage values becomes smoother and kind of symmetrical especially in the graph of 500 sample and 5000 sample. This resembles a normal distribution, which means that with larger sample size, the central tendency become clearer and the data appears to be more normally distributed.
